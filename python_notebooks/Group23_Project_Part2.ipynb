```python
import altair as alt
import pandas as pd
```

```python
brush = alt.selection_interval(encodings=['x'])
```

```python
input_data=pd.read_excel(r'wb_digital_gov_database_jul2020.xlsx', sheet_name = 'DG Projects')
input_data_1 = input_data[['Practice','Status', 'ApprFY','Tot WB ($m)','IDA Co ($m)','Net Co ($m)']]
temp = input_data_1.groupby('ApprFY').size()
print(temp)
```

```python
graph_data = input_data[['Practice','Status', 'ApprFY','Tot WB ($m)','IDA Co ($m)']]
graph_data_1 = graph_data.groupby(['Practice','Status']).size().unstack('Status', fill_value = 0).reset_index()
graph_data_1['Total'] = graph_data_1['Active']+graph_data_1['Closed']+graph_data_1['Pipeline']
graph_data_1 = graph_data_1.sort_values(by = 'Total', ascending=False)
graph_data_1.reset_index(drop=True, inplace = True)
#graph_data_2 = graph_data_1[['Practice','Active','Closed','Pipeline','Total']].reset_index(drop=True)
print(graph_data_1)
```

```python
rect = alt.Chart(input_data_1, title = 'Count of projects vs Year').mark_bar().encode(
x = 'ApprFY:O',
y = 'count:Q'
).transform_aggregate(
    count = 'count(ApprFY)',
    groupby = ['ApprFY']
).add_selection(
    brush
).properties(
    width=380,
    height=250
)
rect
```

```python
#df_pie = data_read.groupby('Status').size().reset_index()
#df_pie.rename(columns = {0:'Count'}, inplace = True)
#print(df_pie)
```

```python
hist = alt.Chart(input_data_1, title = 'Total funding through WB according to the practice').mark_bar().encode(
    x='Practice:O',
    y='sum(Tot WB ($m)):Q',
    color='Status'
).transform_filter(
    brush
).properties(
    width=380,
    height=250
)
#hist
```

```python
scatter = alt.Chart(input_data_1, title = "Corelation between IDA and Net funds in $M").mark_point().encode(
    x='IDA Co ($m):Q',
    y='Net Co ($m):Q',
    #color='Status'
).transform_filter(
    brush
).properties(
    width=380,
    height=250
)
#scatter
```

```python
pie = alt.Chart(input_data_1, title="Percentage of projects exist according to status").mark_arc().encode(
    theta=alt.Theta(field="count", type="quantitative"),
    color=alt.Color(field="Status", type="nominal"),
    #color='Status'
).transform_filter(
    brush
).transform_aggregate(
    count = 'count(Status)',
    groupby = ['Status']
).properties(
    width=380,
    height=250
)
```

```python
dashboard1 = rect | hist 
dashboard2 = scatter | pie
dashboard = alt.vconcat(dashboard1, dashboard2)
dashboard
```

The following dashboard has 4 interactive plots:
1] Bar plot showing number of total projects approved in a year. This plot is helpful to visualize the time-series instance present in the dataset.It is convenient for a user to select range of interested years through x-axis. The rest of the plots will update themselves automatically as per the user selection.
2] Stacked Bar plot showing practice showing sum of total World Bank commitment according to the project status. Each project has 1 of the following status- ACTIVE, CLOSED or PIPELINE.The graph is useful in finding out budgets allocated by World Bank according to the status of the projects for the selected years.
3] The third plot is a scatter plot showing correlation between IDA which is funds allocated by World Bank for e-governance and Net Funds allocated.It can be useful to identify any unusual behaviors in IDA funds.
4] The last plot is a pie chart showing percentages of ACTIVE,CLOSED and PIPELINE projects in given years.This plot is useful for user to priortize their focus according to the status of the project. For instance, if a number of Pipelined projects are greater than a certain threshold then summarized data associated with the projects.

The other contextual data thay we're using is obtained from International Monetary Funds Website "https://data.imf.org/?sk=B83F71E8-61E3-4CF1-8CF3-6D7FE04D0930&sId=1409151240976". The dataset is about reporting of monetary funds reported to the IMF.The dataset contains all  the monetary information such as Loans, Total Assets and information related to deposits and securities. Using this kind of data we can use it to show visualizations such as the countries that have most number of loans or deposits and the amount they pay every month. The dataset includes data of both representations of data of the countries that included data from both SRF(standardized reporting formats) reporting and non-SRF reportings.

We're planning to improve the visualization of the plots to make them more user friendly. For example, two graphs on each line so that users do not need to scroll to view all the plots. In addition to this, we're also planning to add two more interactive plots in the final project draft.
